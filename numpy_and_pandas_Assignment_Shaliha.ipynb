9.
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
● Join them into a single array
● Reshape the result into a 3×3 matrix
● Extract the diagonal elements

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])

joined_array = np.concatenate((a, b, c))
print("Joined array:", joined_array)

reshaped_matrix = joined_array.reshape(3, 3)
print("Reshaped 3x3 matrix:\n", reshaped_matrix)

diagonal_elements = np.diag(reshaped_matrix)
print("Diagonal elements:", diagonal_elements)
```

10. Create a NumPy array of integers from 1 to 20.
● Extract all values that are divisible by 3 or 5.
● Count how many such values exist

```python
H = np.arange(1, 21)
print("Array", H)

divisible = H[(H % 3 == 0) | (H % 5 == 0)]
print("\nValues divisible by 3 or 5:\n", divisible)

count = divisible.size
print("Count of values divisible by 3 or 5:", count)
```
